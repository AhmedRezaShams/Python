```python
mean_bmi = data['bmi'].mean()
data['bmi'] = data['bmi'].fillna(mean_bmi)
data.isnull().sum()
```

```python
data.drop(columns = ['id'], inplace = True)
```

```python
# defining feature matrix(X) and response vector(y) 
X = data.iloc[:,1:-1]
X
```

```python
y = data.iloc[:,-1:]
y
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import OneHotEncoder
#convert categorical to numeric
labelencoder_X=LabelEncoder()
z = X.iloc[:,3]
X.iloc[:,3] = labelencoder_X.fit_transform(z.values.ravel())
X.iloc[:,3]
```

```python
#convert categorical to numeric
labelencoder_X=LabelEncoder()
z = X.iloc[:,4]
X.iloc[:,4] = labelencoder_X.fit_transform(z.values.ravel())
X.iloc[:,4]
```

```python
#convert categorical to numeric
labelencoder_X=LabelEncoder()
z = X.iloc[:,5]
X.iloc[:,5] = labelencoder_X.fit_transform(z.values.ravel())
X
```

```python
#convert categorical to numeric
labelencoder_X=LabelEncoder()
z = X.iloc[:,-1]
X.iloc[:,-1] = labelencoder_X.fit_transform(z.values.ravel())
X
```

```python
#z score normalization
scaller = StandardScaler()
X = scaller.fit_transform(X)
```

```python
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(y)
y = encoder.transform(y)
```

```python
# splitting X and y into training and testing sets  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1) 
```

```python
from keras.utils import np_utils
# convert integers to dummy variables (i.e. one hot encoded)
y_train = np_utils.to_categorical(y_train)
```

```python
# define the keras model
model = Sequential()
model.add(Dense(6, input_dim=9, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(2, activation='softmax'))
```

```python
# compile the keras model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on the dataset
model.fit(X_train, y_train, epochs=100, batch_size=30)

acc = model.evaluate(X_train, y_train)
print("Loss:", acc[0], " Accuracy:", acc[1])

pred = model.predict(X_test)
pred_y = pred.argmax(axis=-1)
```
