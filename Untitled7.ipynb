```python
import pandas as pd
import numpy as np

import plotly.express as px

# Load the CSV file into a pandas dataframe
df = pd.read_csv("/content/Top 50 US Tech Companies 2022 - 2023.csv")

# Print the contents of the dataframe
print(df)
```

```python
print("\nNull Value Information:\n", df.isnull().sum())
```

```python
df.describe()[['Annual Income Tax in 2022-2023 (USD in Billions)', 'Market Cap (USD in Trillions)','Annual Revenue 2022-2023 (USD in Billions)']]
```

```python
#visualization-histogram
fig1 = px.histogram(df, x='Annual Income Tax in 2022-2023 (USD in Billions)')
fig2 = px.histogram(df, x='Market Cap (USD in Trillions)')
fig3 = px.histogram(df, x='Annual Revenue 2022-2023 (USD in Billions)')
fig1.show()
fig2.show()
fig3.show()
```
